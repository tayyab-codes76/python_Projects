># 1. ATM Simulation
* Start with balance 1000
* Show menu repeatedly:
    * Check balnce
    * Deposit
    * Withdraw
    * Exit
* Keep eunning until user chooses Exit

```python
balance = 1000
print("-"*40)
print("Welcome to TY ATM machine!")
print("-"*40)
while True:
    print("Selecet the option one of them:")
    print(
        "1. Check Balance\n"
        "2. Deposit\n"
        "3. Withdraw\n"
        "4. Exit"
    )
    print("-"*40)
    selected_optiion = input()
    if selected_optiion == "Check Balance":
        print(f"You're balance is: {balance}")
        print("-"*40)
    elif selected_optiion == "Deposit":
        print("Enter the amount you wat to Deposit: ")
        amount = int(input())
        total_amount = balance + amount
        print(f"you're balance was:{balance}. Now you Deposit an amount: {amount}. So, you're total balance is now: {total_amount}")
        print("-"*40)
    elif selected_optiion == "Withdraw":
        print("Enter the amount which you want to withdraw: ")
        withdraw_amount = int(input())
        if withdraw_amount > balance:
            print(f"You're balance is: {balance}. You enter the high amount from you're balance. Please try again.")
            break
        else:
            print(f"you're balance was:{balance}. Now you withdraw an amount: {withdraw_amount}. So, you're total remaining balance is now: {balance - withdraw_amount}")
            print("-"*40)
    elif  selected_optiion == "Exit":
        break
    else:
        print("Please entered a correct option!")  
```

># 2. Number Guessing Game
* Set a secret number
* User keeps guessing
* Tell:
    * Too high
    * Too low
* Count number attempts
* Stop when correct
* Print attempts at the end

```python
secret_number = 45
print("-"*40)
print("Welcome to the number guessing game!")
print("-"*40)
i = 0
while True:
    print("-"*40)
    print("Enter a number for guessing our secret number: ")
    print("-"*40)
    guessing_number = int(input())
   
    if guessing_number > secret_number:
        print(f"Ooops! You entered the number {guessing_number}. \nWhich is high from our secret number. Try again")
        i+=1
    elif guessing_number < secret_number:
        print(f"Ooops! You entered the number {guessing_number}.\n Which is low from our secret number. Try again")
        i+= 1
    # elif guessing_number != secret_number:
    #     i += 1
    elif guessing_number == secret_number:
        print(f"Congratulations! You entered the number {guessing_number}.\n Which is the equel of our secret number.")
        i += 1
        break
print("-"*40)
print(f"You are total guessing attemt is: {i}")
```

```python
def students_mangment_system():
    # ===============================
    # 🎯 Welcome Section
    # ===============================
    print("=" * 38)
    print("🎓 Welcome to Student Management System")
    print("=" * 38)

    # Main storage for all students
    students = []

    # ===============================
    # 🔁 Main Menu Loop
    # ===============================
    while True:

        print("\n📋 Select an option (Enter number):")
        print("=" * 38)
        print("1. ➕ Add Student")
        print("2. 📄 View All Students")
        print("3. 🔍 Search Student")
        print("4. ✏️ Update Student")
        print("5. ❌ Delete Student")
        print("6. 🚪 Exit")
        print("=" * 38)

        select = input("Your choice: ")

        # ===============================
        # ➕ Add Student
        # ===============================
        if select == "1":
            student_dic = {}   # New dictionary for each student
            student_marks = []  # List to store marks

            print("\n" + "=" * 40)
            print("📥 Add New Student Data")

            # Taking input
            name = input("Enter student name: ")
            age = int(input("Enter student age: "))
            clas = input("Enter student class: ")

            # Taking marks input
            for i in range(1, 4):
                marks = int(input(f"Enter marks for subject {i}: "))
                student_marks.append(marks)

            # Calculations
            total_marks = sum(student_marks)
            average_marks = round(total_marks / 3, 2)

            # Storing data in dictionary
            student_dic["name"] = name
            student_dic["age"] = age
            student_dic["Class"] = clas
            student_dic["Student Marks"] = student_marks
            student_dic["Total marks"] = total_marks
            student_dic["Average marks"] = average_marks

            # Adding to main list
            students.append(student_dic)

            print("✅ Student added successfully!")
            print("📌 Current Students:", students)

        # ===============================
        # 📄 View Students
        # ===============================
        elif select == "2":
            print("\n📄 All Students Data:")
            if students:
                for s in students:
                    print(s)
            else:
                print("⚠️ No student data available.")

        # ===============================
        # 🔍 Search Student
        # ===============================
        elif select == "3":
            print("\n🔍 Search Student")
            name = input("Enter student name: ").lower()

            for student in students:
                if student["name"].lower() == name:
                    print("✅ Student found:", student)
                    break
            else:
                print("❌ Student not found")

            print("=" * 40)

        # ===============================
        # ✏️ Update Student
        # ===============================
        elif select == "4":
            print("\n" + "=" * 40)
            print("✏️ Update Student Data")

            name = input("Enter student name to update: ").lower()
            found = False

            for student in students:
                if student["name"].lower() == name:
                    found = True
                    print("✅ Student found:", student)

                    change = input(
                        "What do you want to change (name, age, classe, marks)? "
                    ).lower()

                    if change == "name":
                        new_name = input("Enter new name: ").lower()
                        student["name"] = new_name

                    elif change == "age":
                        new_age = int(input("Enter new age: "))
                        student["age"] = new_age

                    elif change == "classe":
                        new_class = input("Enter new class: ")
                        student["classe"] = new_class

                    elif change == "marks":
                        new_marks = []
                        print("Enter new marks:")
                        for i in range(1, 4):
                            m = int(input(f"Subject {i}: "))
                            new_marks.append(m)
                        student["marks"] = new_marks

                    else:
                        print("⚠️ Invalid option! Choose: name, age, classe, marks")

                    print("✅ Student updated successfully!")
                    break

            if not found:
                print("❌ Student not found!")

            print("=" * 40)

        # ===============================
        # ❌ Delete Student
        # ===============================
        elif select == "5":
            print("\n" + "=" * 40)
            print("❌ Delete Student")

            name = input("Enter student name to delete: ").lower()
            found = False

            for student in students:
                if student["name"].lower() == name:
                    found = True
                    print("⚠️ Student found:", student)

                    confirm = input("Are you sure? (yes/no): ").lower()

                    if confirm == "yes":
                        students.remove(student)
                        print("✅ Student deleted successfully!")
                    else:
                        print("❌ Deletion cancelled.")

                    break

            if not found:
                print("❌ Student not found!")

            print("=" * 40)

        # ===============================
        # 🚪 Exit Program
        # ===============================
        elif select == "6":
            print("\n👋 Exiting program... Goodbye!")
            break

        # ===============================
        # ⚠️ Invalid Input
        # ===============================
        else:
            print("⚠️ Wrong input! Please select a valid option (1–6).")


# Run the system
students_mangment_system()
```

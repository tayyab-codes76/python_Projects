># 3. Strong Password Validator

>Task: Keep asking for password until it meets ALL conditions:
* At least 8 characters
* Contains a number
* Contain an uppercase letter

```python
print("-"*40)
print("Creat the password:")
print("-"*40)

print("You're password must followed these requiremets: ")
print("-"*40)
print(
        "1. Password must be atleast 8 characters.\n"
        "2. Password must be contain a number.\n" 
        "3. Password must be contain an uppercase"
)

while True:

    is_digit = False
    is_upper = False

    i = 0

    password = input()
    while i < len(password):
        ch = password[i]
        if ch.isdigit():
            is_digit = True
        if ch.isupper():
            is_upper = True
        i += 1
    
    if len(password) >= 8 and is_digit and is_upper:
        print("-"*40)
        print("Congratulations! You're password is created. don't forget it. ")

        break
    
    else:
       
        if len(password)<8:
            print("-"*40)
            print("You are password is less than 8 characters!\nPlease Try again.")
            print("-"*40)
        elif is_digit != True:
            print("-"*40)
            print("You are password has nat any digit!\nPlease Try again.")
            print("-"*40)
        elif is_upper != True:
            print("-"*40)
            print("You're password has not any uppercase letter!\nPlease Try again.")
            print("-"*40)
```
